```python
pd.DataFrame([['rohan','Male',26],['Manoj','Male',34],['Roopa','Female',22],['Kusum','Female',23]],columns = ['Name','Gender','Age'])
```

```python
# Will create a DataFrame using Nested List

lis = [['Jeevan',100],['Jagruti',98],['Chandana',80],['Roopa',73],['Shravani',84]]
print(lis)

df = pd.DataFrame(data=lis,columns=['Name','Marks'])
df
```

### Adding New column based on Condition 

```python
df.Age.mean()
```

```python
Age_info = []
for i in df.Age:
    if i < df.Age.mean():
        Age_info.append('Younger')
    else:
        Age_info.append('Older')
```

```python
df['Age_range'] = Age_info
```

```python
df.head()
```

### Alternative way to perform same opeartion 

```python
df['Age_range_Lambda'] = df.Age.apply(lambda x: 'Younger'  if x < df['Age'].mean() else 'Older')
```
